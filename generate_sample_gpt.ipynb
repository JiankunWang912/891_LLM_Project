```python
import openai
import os
from openai import OpenAI 
import time 

os.environ['OPENAI_API_KEY'] = '' #FILL IN YOUR OWN HERE
```

```python
# file = open("Adv_cancer_train_70_part_2.csv","r")
file = open("Adv_cancer_train_70_part_1.csv","r")
print("Output of Readlines after appending")
contents = []
cnt = 0
for line in file.readlines():
    if len(line) > 10:
        cnt += 1
        contents.append(line)
        word_num = len(line.split(' '))
        print(f'line {cnt}: word_num: {word_num}')

file.close()

file = open("Adv_cancer_train_70_part_2.csv","r")
for line in file.readlines():
    if len(line) > 10:
        cnt += 1
        contents.append(line)
        word_num = len(line.split(' '))
        print(f'sample {cnt}: word_num: {word_num}')

file.close()
```

```python
# sample_response_dict2 = {}
model_name = 'gpt-3.5-turbo'  # "gpt-4-turbo-preview"
for sample_idx in range(cnt):
    # responses = []
    text = contents[sample_idx]
    instruction = "This is a discharge summary for a patient. Please act as a healthcare professional and help me rephrase this medical record with the similar number of total words. \n"
    prompt =  instruction + text
    # send request
    
    for r in range(9):
        client = OpenAI()
        completion = client.chat.completions.create(
            model=model_name,
            messages=[
                {"role": "user", "content": prompt}
            ], max_tokens=2048)
        
        res = completion.choices[0].message.content.replace("\n", " ")
        res = res.replace('\n\n', '.').replace('\n', ' ').replace('**', '')
        # responses.append(res)
        # print(f"sample {sample_idx}, request {r}\r", end='') 
        with open(f"generated-samples-{model_name}/sample-{sample_idx}.csv", "a") as file:
            file.write(res+'\n')
        print(f"sample {sample_idx}, request {r} write in\r", end='') 
    print(f"Sample {sample_idx} time: {time.strftime('%l:%M%p  on %b %d, %Y')}")
    #sample_response_dict2[sample_idx] = responses
```
